```python
country = response.json()['Countries']
```

```python
country[0]
```

```python
df = pd.io.json.json_normalize(country[0])
```

```python
df
```

```python
for countries in country[1:]:
    df = df.append(pd.io.json.json_normalize(country))
```

```python
df['Date'] = pd.to_datetime(df['Date'], utc=True)
```

```python
df.head()
```

```python
client = MongoClient()
db = client['covid']
collection = db['covid_cases']
```

```python
collection.insert_many(df.to_dict('records'))
```

```python
collection.find_one()
```

```python
sort = collection.find({},{"Country":1, "NewConfirmed":1, "TotalConfirmed":1}).sort("NewConfirmed", -1).limit(1000)
```

```python
df2 = pd.io.json.json_normalize(sort)
```

```python
df2
```

```python
sort2 = collection.find({},{"Country":1, "TotalConfirmed":1, "TotalDeaths":1, "NewConfirmed":1, "NewDeaths":1}).sort("TotalConfirmed", -1).limit(1000)
```

```python
df3 = pd.io.json.json_normalize(sort2)
```

```python
df3
```

```python
sort3 = collection.find({'TotalDeaths':{ '$gt': 20000}}).limit(1000)
```

```python
df4 = pd.io.json.json_normalize(sort3)
```
